```python
def return_df(gene_list: list, columns_selected: list, count_table: pd.DataFrame, sample_metadata: dict) -> pd.DataFrame:
    df_list = []
    for gene in gene_list:
        if gene not in count_table.index:
            raise ValueError('Gene not found')
        gene_count_data = count_table.loc[gene]
        print(gene_count_data.loc['Geneid'])
        genesym = gene_count_data.loc['gene symbol']
        for col, val in zip(gene_count_data.index,gene_count_data):
            if col not in columns_selected: continue
            df_list.append([col,gene,genesym,sample_metadata[col]['brainregion'],sample_metadata[col]['type'],val])
    return pd.DataFrame(df_list, columns=['Sample','ENSMBL Gene ID', 'Gene Symbol','Brainregion','Celltype','Genecount'])

return_df(gene_list,columns_selected,count_table,sample_metadata)
```

```python
def plot_gene_correlation(gene1: dict, gene1_str: str, gene2: dict, gene2_str: str, relative: bool = True):
    color = []
    shape = []
    count_y = []
    count_x = []
    for brainregion in gene1:
        for key in gene1[brainregion]:
            for val1, val2 in zip(gene1[brainregion][key],gene2[brainregion][key]):
                total = val1 + val2 if relative else 1
                count_y.append(val1/total)
                count_x.append(val2/total)
                color.append(key)
                shape.append(brainregion)
    df = pd.DataFrame({gene1_str: count_y, gene2_str: count_x, 'color': color, 'shape': shape})
    fig, ax = plt.subplots(1,1)
    fig.set_size_inches(10,10)
    sns.scatterplot(df,y=gene1_str,x=gene2_str,style='shape',hue='color',axes=ax, s=100)
    if relative:
        plt.ylim(0,1)
        plt.xlim(0,1)
    else:  
        plt.ylim(0,max(count_y)+1000)
        plt.xlim(0,max(count_x)+1000)
    plt.show()
```
